```python
# importing libraries
import pandas as pd
```

```python
# reading files
df = pd.read_csv('employee.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
grp_df = df.groupby(['Department','EducationField'])[['Age','DailyRate','MonthlyIncome']].agg('mean')
```

```python
grp_df
```

```python
# highlighting the max and min value
grp_df.style.highlight_max(color = 'red').highlight_min(color='lightgreen')
```

```python
def above_thousand(x):
    color = 'red' if x>1000 else 'green'
    return 'color: %s' % color
```

```python
df[['DailyRate']].style.applymap(above_thousand)
```
