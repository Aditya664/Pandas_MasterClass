```python
# importing libraries
import pandas as pd
```

```python
# reading data
df = pd.read_csv('employee.csv')
```

```python
# head of the data
pd.set_option('max_columns', 35)
df.head()
```

```python
# grouped the data by education field
grp_data = df.groupby(['EducationField'])[['Age','DailyRate','HourlyRate','MonthlyIncome']].agg('mean')
```

```python
grp_data
```

### Pandas Styling

```python
# highlighting the maximum value in each column
grp_data.style.highlight_max(color = 'darkgreen')
```

```python
# using method chains to highlight the max and min value
grp_data.style.highlight_max(color = 'darkgreen').highlight_min(color = 'yellow')
```

```python
grp_data.style.background_gradient(cmap = 'Greens')
```

```python
# method chains to hide the index as well
grp_data.style.background_gradient(cmap = 'Greens').hide_index()
```
