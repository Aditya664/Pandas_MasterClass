## Cross Tabulation
Cross tabulation is a statistical tool that is used to analyze the categorical data.
The crosstab() function of pyhton is used to compute a simple cross tabulation of two or more factors.


### Example 1

```python
import numpy as np
import pandas as pd
# creating arrays
a = np.array(["foo", "foo", "foo", "foo", "bar", "bar",
              "bar", "bar", "foo", "foo", "foo"], dtype=object)
b = np.array(["one", "one", "one", "two", "one", "one",
              "one", "two", "two", "two", "one"], dtype=object)
c = np.array(["dull", "dull", "shiny", "dull", "dull", "shiny",
              "shiny", "dull", "shiny", "shiny", "shiny"],
             dtype=object)
```

```python
# performing crosstabulation
pd.crosstab(a, [b, c], rownames=['a'], colnames=['b', 'c'])
```
