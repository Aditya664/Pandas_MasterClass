## Binning
Binning method is used to smoothing data or handle noisy data. 

It is a way to group a number of more or less continuous values into smaller number of bins.



```python
# importing all required libraries
import pandas as pd
import numpy as np
```

```python
# Discovers the same bins, but assign them specific labels
pd.cut(np.array([1, 7, 5, 4, 6, 3]),3, labels=["bad", "medium", "good"])
```

### Example 2

```python
# reading data
df = pd.read_csv('employee.csv')
df.head(3)
```

```python
pd.cut(df['DistanceFromHome'],3, labels = ['Near', 'Far', 'Extremely Far'])
```

```python
# value counts for distance from home column
df['DistanceFromHome'].value_counts()
```

```python
#Defining the function which will bin the numerical DistanceFromHome variable into Categories

def distance_from_home(x):
    if x <= 2:
        return 'Near'
    elif x > 2 and x <=5:
        return 'Far'
    elif x >5:
        return 'Extremely Far'

df['DistanceFromHome'] = df['DistanceFromHome'].apply(distance_from_home)
df['DistanceFromHome'].value_counts()
```
